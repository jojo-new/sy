```python
from sklearn.model_selection import train_test_split

X = df.loc[:, df.columns != 'MEDV']
y = df.loc[:, df.columns == 'MEDV']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler #scales the features in bet 0 1nd 1
mms = MinMaxScaler()
mms.fit(X_train) # learn min max from traning data
X_train = mms.transform(X_train) # scale both taring and testing data acc.
X_test = mms.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential # linear stack of the layer
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(128, input_shape=(13, ), activation='relu', name='dense_1'))
model.add(Dense(64, activation='relu', name='dense_2'))
model.add(Dense(1, activation='linear', name='dense_output'))

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=100, validation_split=0.05, verbose = 1)
```

```python
mse, mae = model.evaluate(X_test, y_test)

print('MSE: ', mse)
print('MAE: ', mae)
```
